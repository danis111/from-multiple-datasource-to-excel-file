```python
import os
current_folder = os.getcwd()
print(current_folder)
files = os.listdir(current_folder)
# files
#This is current directory we're working ond

export(files)
```

```python

# those are files that's gonna be imported to DataFrame

def export(files):
    
    import pandas as pd
    sales = pd.DataFrame()
    
    for file in files:
        if  os.path.join(file).split(".")[-1]=="xlsx":
            df = pd.read_excel(file)
            sales = sales.append(df,ignore_index=True)
            
        
        elif os.path.join(file).split(".")[-1]=="csv":
            df = pd.read_csv(file)
            sales = sales.append(df,ignore_index=True)
            
        elif os.path.join(file).split(".")[-1]=="json":
            df = pd.read_json(file,lines=True)
            sales = sales.append(df,ignore_index=True)
        
        else:
            continue
            
            
            
    return sales     
        
```
